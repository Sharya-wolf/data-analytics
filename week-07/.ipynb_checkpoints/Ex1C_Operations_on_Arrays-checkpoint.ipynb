# Exercise 1.C Performing operations on arrays

```python
import numpy as np
```

## Sample Arrays

```python
r_1 = np.random.randint(1,21, size=2)
r_2 = np.random.randint(1,21, size=(1,2))
r_3 = np.random.random(size=(1,2))
r_4 = np.random.randint(20,101, size=(4,5))
r_5 = np.random.random(size=(4,5))

print(r_1)
print(r_2)
print(r_3)
print(r_4)
print(r_5)
```

## Permutations

```python
print("Original r_4:")
print(r_4)

np.random.shuffle(r_4)

print("\nShuffled r_4:")
print(r_4)
```

```python
# After using random shuffle, the rows are shuffled in place. It only rearranges the 
# order of the rows but not the elements inside them.
```

```python
new_shuffle = np.random.permutation(r_4)
print("new shuffle:")
print(new_shuffle)

print("\nr_4 (original after permutation):")
print(r_4)
```

You know when to use them, depending on whether you want to change the original data or keep it unchanged.
