# Exercise 2.A Using built-in functions
## Lab 2. Parsing strings - part codes

```python
supplier1 = 'ACME:123-L'
supplier2 = 'DI:12345-M'
supplier3 = 'ACE:1-12'
```

```python
def get_supplier_code(string):
    colon = string.find(":")
    supplier = string[:colon]
    return supplier
```

```python
print(get_supplier_code(supplier1))
print(get_supplier_code(supplier2))
print(get_supplier_code(supplier3))
```

```python
def get_product_num(string):
    colon = string.find(":")
    hyphen = string.find("-")
    product_code = string[colon+1:hyphen]
    return product_code
```

```python
print(get_product_num(supplier1))
print(get_product_num(supplier2))
print(get_product_num(supplier3))
```

```python
def get_size(string):
    hyphen = string.find("-")
    size = string[hyphen+1:]
    return size
```

```python
print(get_size(supplier1))
print(get_size(supplier2))
print(get_size(supplier3))
```

```python
print(f"For {supplier1}, the supplier is {get_supplier_code(supplier1)}.")
print(f"Part number {get_product_num(supplier1)} is available in size {get_size(supplier1)}.")
```

```python
print(f"For {supplier2}, the supplier is {get_supplier_code(supplier2)}.")
print(f"Part number {get_product_num(supplier2)} is available in size {get_size(supplier2)}.")
```

```python
print(f"For {supplier3}, the supplier is {get_supplier_code(supplier3)}.")
print(f"Part number {get_product_num(supplier3)} is available in size {get_size(supplier3)}.")
```
