## Lab 1. Working with names - parsing strings

```python
test_name = 'John Doe'
```

```python
def parse_and_display(name):
    for i in range(len(name)):
     if name[i] == ' ':
         first = name[:i]
         last = name[i+1:]
         print(f"Full name: {name}")
         print(f"First name: {first}")
         print(f"Last name: {last}")
```

```python
parse_and_display(test_name)
```

```python
def parse_and_display2(name):
    if name.count(' ') ==0:
        first = name
        print(f"First name: {first}")
    elif name.count(' ') ==1:
        for i in range(len(name)):
            if name[i] == ' ':
                first = name[:i]
                last = name[i+1:]
        print(f"First name: {first}")
        print(f"Last name: {last}")
    elif name.count(' ') ==2:
        spaces = []
        for i in range(len(name)):
            if name[i] == ' ':
                spaces.append(i)

        first = name[:spaces[0]]
        middle = name[spaces[0]+1:spaces[1]]
        last = name[spaces[1]+1:]

        print(f"First Name: {first}")
        print(f"Middle Name: {middle}")
        print(f"Last Name: {last}")
```

```python
parse_and_display2('Lorde')
```

```python
parse_and_display2('Billie Eilish')
```

```python
parse_and_display2('Megan Thee Stallion')
```

```python
parse_and_display2("Sha'Rya Sha'Kira Weaver")
```
