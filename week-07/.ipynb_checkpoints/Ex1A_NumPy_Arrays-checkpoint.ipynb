# Exercise 1.A NumPy: Manipulating Arrays

```python
import numpy as np
```

## Looping Through Arrays

```python
arrA = np.array([
    [11, 12, 13],
    [14, 15, 16]
])
print(arrA)
```

```python
arrA =np.array([
    [11, 12, 13],
    [14, 15, 16]
])

for x in arrA:
    print(x * 2)
```

```python
arrA = np.array([
    [11, 12, 13],
    [14, 15, 16]
])

for x in arrA:
    for y in x:
        print(y * 2)
```

The difference between the output of the first loop and second is that the first loop gives back each row doubled while the second loop gives back the numbers indivisually from looping through each element inside the row.

```python
arrA = np.array([
    [11, 12, 13],
    [14, 15, 16]
])

for x in arrA:
    for y in x:
        loop_result = x * 2
        print(loop_result)
# The result is different because we are looping through each element, but the calculation is done
# through each row instead. So now it will print the same row result instead of changing with each
# element(number).
```

## Joining and Stacking Arrays

```python
joined_array_1 = np.concatenate((arrA, arrA), axis=0)
print(joined_array_1)

# axis=0 means you join row by row (vertical stacking), so you get the same array placed 
# underneath itself.
```

```python
joined_array_2 = np.concatenate((arrA, arrA), axis=1)
print(joined_array_2)
```

```python
stacked_array_1 = np.concatenate((joined_array_1, joined_array_1))
print(stacked_array_1)
```

```python
stacked_array_2 = np.concatenate((joined_array_1, joined_array_2))
print(stacked_array_2)
# The error message I get is a ValueError. All arrays must match except for the concatenation.
# Array 1 has 3 columns while Array 2 has 6 columns.
```

```python
another_joined_array = np.concatenate ((joined_array_1, arrA))
print(another_joined_array)
```

```python
another_2 = np.concatenate((joined_array_1, arrA), axis=1)
# When you run this code, it will give you a ValueError.
# To put it in my own words, it is because they do not have the same number of rows. 
# joined_array has 4 rows while arrA has 2 rows, so they can't be joined on axis=1.
```

```python
Vari1 = np.hstack((arrA, arrA))
# This is a horizantal stack. As in the name it makes the stack horizantal by adding them 
# side-by-side.
Vari2 = np.vstack((arrA, arrA))
# This is a vertical stack. It makes the stacks vertical by adding rows onn top of each other.
Vari3 = np.dstack((arrA, arrA))
# This is a depth stack. It stacks the array along a third deminsional.

print(Vari1)
print()

print(Vari2)
print()

print(Vari3)
```

## Splitting Arrays

```python
splitsky = np.split(joined_array_1, 2)
print(splitsky)
# This returns a list with 2 seperate arrays. Using still the same dimension which is 
# 2-dimension with 2 columns and 3 rows.
```

```python
spliterino = np.split(joined_array_1, 3)
print(spliterino)
# This will give an error because the array has on four rows and can't evenly be split into
# three. 
```

## Searching and Replacing Array Values

```python
result_1 = np.where(arrA % 2 == 0, arrA, "odd")
print(result_1)
```

## Sorting and Filtering Arrays

```python
arrB = np.array([
    [7,6],
    [13, 5],
    [7, 3]
])

arrB_sorted = np.sort(arrB)
print(arrB_sorted)
```

```python
print(arrB)

z = arrB < 7
arrC = arrB[z]

print(arrC)
```

```python
print(arrB)

y = arrB >= 7
arrD = arrB[y]

print(arrD)
```
