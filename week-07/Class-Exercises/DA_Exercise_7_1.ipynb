## Import NumPy & Creating Revenue Array
Importing numpy and creating 1-D array of weekly revenue figures

```python
import numpy as np
```

```python
revenue = np.array([52_400.0, 38_750.0, 61_200.0, 47_900.0, 55_650.0])
print(f"Array : {revenue}")
print(f"dtype : {revenue.dtype}") 
print(f"Shape : {revenue.shape}")
```

## Statistics For The Revenue

```python
Total_revenue = np.sum(revenue)
Mean_revenue = np.mean(revenue)
Median_revenue = np.median(revenue)
Std_deviation = np.std(revenue)
Min_revenue = np.min(revenue)
Max_revenue = np.max(revenue)
```

```python
print(f"Total revenue : ${(Total_revenue):>12,.2f}")
```

```python
print(f"Mean revenue : ${(Mean_revenue):>12,.2f}")
```

```python
print(f"Median revenue : ${(Median_revenue):>12,.2f}")
```

```python
print(f"Standard Deviation revenue : ${(Std_deviation):>12,.2f}")
```

```python
print(f"Min revenue : ${(Min_revenue):>12,.2f}")
```

```python
print(f"Max revenue : ${(Max_revenue):>12,.2f}")
```

## Element-wise Arithmetic
Applying 5% performance bonus to every region using multiplication

```python
revenue_bonus = revenue * 1.05 # This multiplies every element by 1.05
bonus_amount = revenue_bonus - revenue # gives the exact bonus amount. Element-wise subtraction
```

```python
print(f"Original Revenue  :", revenue)
print(f"Revenue + 5% Bonus:", revenue_bonus)
print("Bonus Amount       :", bonus_amount)
```
