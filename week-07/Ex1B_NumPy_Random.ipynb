## Generating Random Numbers

```python
r_1 = np.random.randint(1,21, size=2)
print(r_1)
```

```python
r_2 = np.random.randint(1,21, size=(1,2))
print(r_2)
```

```python
r_3 = np.random.random(size=(2,1))
print(r_3)
```

```python
r_4 = np.random.randint(20,101, size=(4,5))
print(r_4)
```

```python
r_5 = np.random.random(size=(4,5))
print(r_5)
```

## Generating Random Choices from an Array

```python
students = np.array([ 
    "Sharleen", "Vesna", "Jordan", "Leon", 
    "Lilly", "Demy", "Yanuris", "Malaika", 
    "Kasaiha", "Jonathan", "Marenza", "Dania"
])
```

```python
call_on = np.random.choice(students)
print(call_on)
```

```python
breakout_groups = np.random.choice(students, size=(3,4))
print(breakout_groups)
# This can be used to assign breakout rooms, but it's not ideal because students can be 
# chosen twice or not at all. It may be random, but random can also give duplicates of
# one value.
```

```python
p_values = [0.30, 0.15, 0.10, 0.10,
            0.08, 0.07, 0.05, 0.05,
            0.04, 0.03, 0.03, 0.00]
# Probability values for each student:
# Sharleen = 30% chance
# Vrsna = 15% chance
# Jordan = 10% chance
# Leon = 10% chance
# Lilly = 8% chance
# Demy = 7% chance
# Yanuris = 5% chance
# Malaika = 5% chance
# Kasaiha = 4% chance
# Jonathan = 3% chance
# Marenza = 3% chance
# Dania = 0% chance (will never be selected)

call_on_weighted = np.random.choice(students, size=5, p=p_values)
print(call_on_weighted)
```

## Generating a Normal Distribution

```python
bell_curve = np.random.normal(loc=10, scale=5, size=(4, 5))
print(bell_curve)
```
