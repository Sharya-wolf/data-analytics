## Lab 2 Working with datetime functions

```python
import datetime
```

# Today

```python
Today = datetime.datetime.now()
```

```python
# .strftime()
# .strftime() means string format time put together. It's to make the datetime into a readable style.
print("Today is", Today.strftime("%A, %B %d, %Y"))
# %A means day of week, %B means month, %d means day and %Y means year.
# Be careful with putting commas, as you can see B and d don't have any commas between them 
# because the print will add the comma between them. As well as being careful not to capitalize
# the d, or your output will change.
```

# My Birthday

```python
Birthday = datetime.date(2005, 10, 23) # By inputting the number we are storing the date
```

```python
print("My birthday is", Birthday.strftime("%m/%d/%y"))
# As you can see, when you put the lowercase letters, it uses numbers instead of the words.
# As for the year, it gets shortened to only using the last two numbers of your year.
```

# Next 90 Days

```python
ninety_d = Today + datetime.timedelta(days=90)
print("90 days from today is", ninety_d.strftime("%B %d, %Y"))
```

# Time for Dinner

```python
dinner_time = datetime.time(18, 20)
```

```python
print("Let's meet for dinner at", dinner_time.strftime("%I:%M %p"))
```
